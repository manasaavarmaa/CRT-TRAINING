```python
!pip install transformers

from transformers import pipeline

def summarize_text(text):
  """Summarizes a given text using a pre-trained Hugging Face model."""
  summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
  summary = summarizer(text, max_length=130, min_length=30, do_sample=False)
  return summary[0]["summary_text"]

# Example usage
text_to_summarize = """
The quick brown fox jumps over the lazy dog. This is a simple sentence.
Another sentence to add some more text.  Here's a bit more text.
The dog is not lazy. He is just playing. The fox is very clever.
"""
summary = summarize_text(text_to_summarize)
print("Summary:", summary)
```
